#### Create two scatter plots comparing two variables and provide your analysis on correlation and causation. Remember, covariance, Pearson’s correlation, and Non-Linear Relationships should also be considered during your analysis (Chapter 7).

```python
from scipy.stats import pearsonr

year_variable = 'Year'
lat_variable = 'lat'

plt.figure(figsize=(12, 8))
sns.scatterplot(x=year_variable, y=lat_variable, data=df)
plt.title(f'Scatter Plot: {year_variable} vs. {lat_variable}')
plt.xlabel(year_variable)
plt.ylabel(lat_variable)
plt.show()

plt.figure(figsize=(12, 8))
sns.scatterplot(x=lat_variable, y=year_variable, data=df)
plt.title(f'Scatter Plot: {lat_variable} vs. {year_variable}')
plt.xlabel(lat_variable)
plt.ylabel(year_variable)
plt.show()

cov_matrix = df[[year_variable, lat_variable]].cov()

pearson_corr, _ = pearsonr(df[year_variable], df[lat_variable])

print(f"Covariance Matrix:\n{cov_matrix}")
print(f"Pearson's Correlation Coefficient between {year_variable} and {lat_variable}: {pearson_corr}")
```

> The correlation coefficient of 0.19 between 'Year' and 'lat' suggests a weak positive correlation, indicating that there is some tendency for latitude values to increase with the year. However, correlation alone does not imply causation. The weak correlation may be influenced by various factors, and it's important to conduct further analyses and consider additional variables to establish any causal relationship. Causation requires a more in-depth understanding of the underlying mechanisms and potential confounding variables that could be influencing both 'Year' and 'lat.'

#### Conduct a test on your hypothesis using one of the methods covered in Chapter 9

> The code below conducts a Pearson correlation test between the 'Year' and 'lat' variables in a DataFrame, calculates the correlation coefficient and p-value, and determines whether the correlation is statistically significant based on a chosen significance level (alpha).

```python
import pandas as pd
import scipy.stats as stats

variable1 = 'Year'
variable2 = 'lat'
data = df[[variable1, variable2]].dropna()
corr_coef, p_value = stats.pearsonr(data[variable1], data[variable2])
print(f'Correlation Coefficient: {corr_coef}')
print(f'P-value: {p_value}')

alpha = 0.05
if p_value < alpha:
    print('The correlation is statistically significant.')
else:
    print('There is no significant correlation.')
```

> The correlation coefficient of 0.185 suggests a positive but relatively weak linear relationship between the 'Year' and 'lat' variables. The extremely low p-value (1.39e-41) indicates that this correlation is statistically significant, suggesting that it is unlikely to have occurred by random chance.

```python
import pandas as pd
import statsmodels.api as sm

year_variable = 'Year'
lat_variable = 'lat'
X = sm.add_constant(df[year_variable])
y = df[lat_variable]
model = sm.OLS(y, X).fit()
print(model.summary())
```
